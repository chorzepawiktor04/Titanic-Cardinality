```python
import pandas as pd

columns = ['pclass', 'survived', 'name', 'sex', 'age', 'sibsp', 'parch', 'ticket', 'fare', 'cabin', 'embarked', 'boat', 'body', 'homeDest']
data=pd.read_csv('dane.csv', skiprows=17, sep=',', names=columns, na_values='?')

data
```

```python
print('Liczba etykiet zmiennej survived:{}'.format(len(data.survived.unique())))
```

```python
print('Liczba etykiet zmiennej age:{}'.format(len(data.age.unique())))
print('Liczba etykiet zmiennej sibsp:{}'.format(len(data.sibsp.unique())))
print('Liczba etykiet zmiennej parch:{}'.format(len(data.parch.unique())))
print('Liczba etykiet zmiennej cabin:{}'.format(len(data.cabin.unique())))
print('Liczba etykiet zmiennej ticket:{}'.format(len(data.ticket.unique())))
print('Liczba etykiet zmiennej body:{}'.format(len(data.body.unique())))
print('Liczba etykiet zmiennej boat:{}'.format(len(data.boat.unique())))
print('Liczba etykiet zmiennej embarked:{}'.format(len(data.embarked.unique())))
print('Liczba etykiet zmiennej homeDest:{}'.format(len(data.homeDest.unique())))
```

```python
print('Liczba pasażerów :{}'.format(len(data.index.unique())))
```

Mała kardynalność: survived, sex, sibsp, parch, embarked
Duża kardynalność: age, body, homeDest, ticket, cabin
Zmienna boat jest gdzieś pomiędzy, jednak bardziej się skłaniam do zaliczenia jej do dużej kardynalności.

```python
import numpy as np

cabinUnique = np.unique(data['cabin'].dropna())
print(cabinUnique)
```

```python
data['CabinReduced'] = data['cabin'].dropna().astype(str).str[0]
print(data[['cabin', 'CabinReduced']].head(20))
```

```python
print('Liczba etykiet zmiennej CabinReduced:{}'.format(len(data.CabinReduced.unique())))
```

zredukowano kardynalność o nieco ponad 95%

W przypadku zmiennej cabin taka redukcja nie zmienia praktycznie nic w dalszej analizie - możemy się domyślać ze pierwsza litera jest nośnikiem istotnych informacji w tej kolumnie - prawdopodobnie był podział na sekcje A,B,C..., a to co stało za literą było jedynie numeracją kabin -> nie mamy informacji jak były one rozmieszczone w zależności od numeru więc nie możemy nic z taką informacją zrobić.
Cabin jest tak naprawdę jedyną kolumną w tym zbiorze, dla której można dokonać takiej redukcji kardynalności, bez żadnego negatywnego wpływu na dalszą analizę.
